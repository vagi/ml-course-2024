**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
# Ваш код здесь
print(data.groupby(by=['race', 'sex'])['age'].describe())
print(f"Відповідь: максимальний вік чоловіка раси Amer-Indian-Eskimo = \
{data.groupby(by=['race', 'sex'])['age'].describe().loc[['Amer-Indian-Eskimo'], 'max'][1]} роки.")
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Ваш код здесь
married_men_percent = data[(data['salary'] == '>50K') & (data['sex'] == 'Male')]['marital-status'].value_counts(normalize=True).\
loc[['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']].sum()
print(f"Відповідь: частка заробляючих багато (>50K) більше серед одружених чоловіків, їх {round(married_men_percent *100, 2)} %.")
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Ваш код здесь
max_hours_per_week = data['hours-per-week'].max()
print(f"Відповідь: Максимальна кількість годин на тиждень, що працює людина, складає: {max_hours_per_week} годин")

number_people_max_hours = data[data['hours-per-week'] == max_hours_per_week].shape[0]
print(f"Таку кількість годин працює: {number_people_max_hours} людей")

high_salary_receivers = data[(data['hours-per-week'] == max_hours_per_week) & (data['salary'] == '>50K')].shape[0]
high_salary_percent = high_salary_receivers / number_people_max_hours *100
print(f"Серед людей, що працюють максимальний час на тиждень, люди що заробляють \
багато, складають {round(high_salary_percent, 2)}%")
```
